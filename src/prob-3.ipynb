Menentukan setiap kolom numerik berdistribusi normal atau tidak. Gunakan normality test
yang dikaitkan dengan histogram plot.


```python
import pandas as pd
import scipy.stats as scp
import seaborn as sns
import matplotlib.pyplot as plt

# All columns contain numeric data
df = pd.read_csv('../dataset/anggur.csv')

# Drop target column
df = df.drop(['quality'], axis=1)
```

```python
df.head()
```

```python
# Normality test using Shapiro-Wilk method
# The Shapiro-Wilk test is a statistical test of the hypothesis that the distribution of the data as a whole deviates from a comparable normal distribution. If the test is non-significant (p>. 05) it tells us that the distribution of the sample is not significantly different from a normal distribution.

for column in df.columns:
    stat, p = scp.shapiro(df[column])
    alpha = 0.05

    print('p-value untuk kolom ' + str(column) + ' adalah ' + str(p))
    if p > alpha :
        print('Kolom ' + str(column) + ' berdistribusi normal')
    else :
        print('Kolom ' + str(column) + ' tidak berdistribusi normal')

    sns.displot(data = df[column], kde = True)
    plt.show()
    print()
    
```
